```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv("DentistryPGRs.csv")
```

```python
df.head()
```

```python
plt.figure(figsize = (8,6))
sns.set_context("paper", font_scale=1.3)
ax = sns.displot(df['Main supervisor'])
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
```

```python
fig2 = plt.figure(figsize = (8,6))
sns.set_context("paper", font_scale=1.8)
ax = sns.displot(df['Type'])
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
fig2.savefig('PGRsnum.pdf')
```

```python
plt.figure(figsize = (8,6))
sns.set_context("paper", font_scale=1.8)
ax = sns.displot(df['Fee status'])
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
```

```python
#fig1=plt.figure(figsize=(6,4))
#plt.tight_layout()
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.6)
sns.swarmplot(x="Fee status", y="Type", data=df, hue="Mode", palette='prism')
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
plt.legend(bbox_to_anchor=(1, 1))
#fig1.savefig("PGRs.png") #output figure was cropped...
```

```python
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.8)
sns.swarmplot(x="Mode", y="Type", data=df)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
```

```python
sns.set_style("darkgrid")
sns.set_context("paper", font_scale=1.8)
sns.stripplot(x='Fee status', y='Type', data=df)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
```
